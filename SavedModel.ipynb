```python
import tensorflow.compat.v1 as tf
import numpy as np
from pandas.io.parsers import read_csv

tf.compat.v1.disable_eager_execution()

model = tf.global_variables_initializer() # 모델 초기화

data = read_csv('pricedata.csv', sep = ',')

xy = np.array(data, dtype = np.float32)
```

```python
print(xy)
```

```python
x_data = xy[ : , 1:-1]# 파이썬 기본 슬라이싱 문법, 변인 네개를 가져옴
y_data = xy[: , [-1] ] # 가장오른쪽의 price만 가져옴 

X= tf.placeholder (tf.float32, shape = [None, 4] )
Y= tf.placeholder (tf.float32, shape = [None, 1] )
W= tf.Variable(tf.random_normal([4,1] ) , name = 'weight') 
b= tf.Variable(tf.random_normal([1]) , name = 'bias') 


```

```python
hypothesis= tf.matmul(X,W)  + b # 가설식 

cost = tf.reduce_mean( tf.square(hypothesis - Y )) # square : 제곱함수

optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.000005)
#학습시키는 데이터의 유형이나 정제된 정도에 따라 학습률인 learning_rate 는 적절히 정해야한다.

train = optimizer.minimize(cost) 
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(100001):
        cost_, hypo_, _ = sess.run( [cost, hypothesis, train ] , feed_dict = {X :  x_data , Y: y_data})
        if step % 500 == 0:
                print( "#" , step, "손실비용: " , cost_) 
                print("- 배추가격: " , hypo_[0] )
                
                
```

```python
#학습된 모델 저장 방법 
saver = tf.train.Saver()

save_path = saver.save(sess,  "./saved.cpkt")
print("학습된 모델을 저장합니다.")

```
